## Task 4.1
The text files `PATIENT.txt`, `STAFF.txt` and `APPOINTMENT.txt` store the comma-separated values for each of the tables in the database.

Write a Python program to read in the data from each file and then store each item of data in the correct place in the database.

Run your program.
<div style="text-align: right">[4]</div>

```python
# Task 4.1
import csv
import sqlite3

def insert_into_db(table: str, file: str):
    try:
        db = sqlite3.connect('./Resources/TASK4/CLINIC.db')
        allowed_tables = ['patient', 'staff', 'appointment']
        if table.lower() not in allowed_tables:
            raise NotImplementedError(f"Cannot use table '{table}'")
        with open('./Resources/TASK4/' + file) as f:
            r = csv.reader(f)
            for row in r:
                placeholders = ', '.join('?' * len(row))
                db.execute(f"INSERT INTO {table} VALUES ({placeholders})", row)
        db.commit()
    finally:
        db.close()
    db.close()
insert_into_db('Patient', "PATIENT.txt")
insert_into_db('Staff', "STAFF.txt")
insert_into_db('Appointment', "APPOINTMENT.txt")
```
